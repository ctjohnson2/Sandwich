```python
# Function to scrape the hyperlinks and volumes for each drug
def get_additional_links(hyperlinksCleanest):
    # List to store all the generated links and volumes from each URL
    all_links = [] # This will store all the generated medication links
    all_volumes=[] # This will hold volumes for all URLs

    # Loop through each URL in the provided list of URLs
    for URL in hyperlinksCleanest:
        print(f"Processing URL: {URL}")
        
        # Initialize the browser (WebDriver)
        browser = uc.Chrome()

        try:
            # Visit the URL
            browser.get(URL)
            browser.maximize_window()
            browser.execute_script("document.body.style.zoom='25%'")  # Apply zoom style
            time.sleep(1)

            # Get the page source and parse it with BeautifulSoup
            soup = BeautifulSoup(browser.page_source, 'html.parser')
            # Convert the page content into a pretty-formatted string (for easier regex matching)
            str_page = soup.prettify()
             # Clean up the string to remove unwanted characters (like quotes and backslashes)
            str_ = str(str_page).replace('"', '').replace('''\\''', "")

            # Use regular expressions to find all instances of "slug" that indicate a medication link
            tmp_list = re.findall("slug\:(.*?)\}", str_)

            print(tmp_list) # Print the list for debugging

            # Initialize temporary lists to store links and volumes for this URL
            new_links = []
            volumes=[]
            # Loop through each item in the list to extract relevant information (link and volume)
            for elem in tmp_list:
                if elem.count('strength') > 0:
                    new_link = "https://www.costplusdrugs.com/medications/" + elem.split(',')[0].replace('u0026', '&')
                    volume = elem.split(',')[1].replace("volume:","").replace(" ","")
                    new_links.append(new_link)
                    volumes.append(volume)
            # Append the new links and volumes from this URL to the respective lists
            all_links.extend(new_links)
            all_volumes.extend(volumes) 

        finally:
            # Close the browser after processing the current URL to free up resources
            browser.quit()

    # Return the final lists containing all generated links and volumes for all URLs
    return all_links,all_volumes

# Call the function and print the result
new_links, volumes = get_additional_links(hyperlinksCleanest)
# Print all the results
print("All generated links:")
print(new_links)
print("All volumes:")
print(volumes)
```

```python
#getting all the hyperlinks for each drug
def scrape(new_links):
    
    # List to store all the generated links and volumes from each URL
    all_dnames=[]
    all_fsqs=[]
    all_cpmfcs=[]

    # Loop through each URL in the provided list
    for URL in new_links:
        print(f"Processing URL: {URL}")
        
        # Initialize the browser (WebDriver)
        browser = uc.Chrome()

        try:
            # Visit the URL
            browser.get(URL)
            time.sleep(1)
            browser.maximize_window()
            browser.execute_script("document.body.style.zoom='25%'")  # Apply zoom style
            #time.sleep(1)
            # Create new links based on the regex matches
            dname = WebDriverWait(browser, 10).until(EC.visibility_of_element_located(
                (By.XPATH, "//*[@id=\"content-area\"]/div/section[1]/h1"))).text
            print(dname)
            fsq = WebDriverWait(browser, 10).until(EC.visibility_of_element_located(
                (By.XPATH, "//*[@id=\"content-area\"]/div/div[1]/section/div/p[1]"))).text
            print(fsq)
            try:
                cpmfc = WebDriverWait(browser, 10).until(EC.visibility_of_element_located(
                    (By.XPATH, "//*[@id=\"content-area\"]/div/div[2]/div/section/div/dl[2]/div[1]/div/div/div/dl/dd"))).text
            except:
                try:
                    WebDriverWait(browser, 10).until(
                        EC.presence_of_element_located((By.XPATH, "//*[@id=\"content-area\"]/div/div[2]/div/section/div/dl[2]/div[1]/div/div/div/dl/dd"))
                    )
                    cpmfc = browser.find_element(By.XPATH, "//*[@id=\"content-area\"]/div/div[2]/div/section/div/dl[2]/div[1]/div/div/div/dl/dd").text
                except:
                    cpmfc = 'NotAvailable'
            print(cpmfc)
            all_dnames.append(dname)
            all_fsqs.append(fsq)
            all_cpmfcs.append(cpmfc)

        finally:
            # Close the browser after processing the URL
            browser.quit()

    # Return the list of all generated links and volumes from all URLs
    return all_dnames,all_fsqs, all_cpmfcs

# Call the function and print the result
dnames, fsqs, cpmfcs = scrape(new_links)
# Print all the results
print("All generated links:")
print(new_links)
print("All volumes:")
print(volumes)
print("All dnames:")
print(dnames)
print("All fsqs:")
print(fsqs)
print("All cpmfcs:")
print(cpmfcs)
```

```python
#zip function combines the lists into a data frame and renaming columns and removing duplicates
output_df = pd.DataFrame(data = zip(dnames,fsqs,volumes,cpmfcs,new_links),columns=['CPDCName','FSQ','Volume','MFRCost','URL']).drop_duplicates()
```

```python
# Extract the dose form (the part before '  ') from the 'FSQ' column and assign it to a new column 'DoseForm'
output_df['DoseForm'] = output_df['FSQ'].apply(lambda x: x.split('  ')[0])
# Extract the strength (the part after '  ') from the 'FSQ' column and assign it to a new column 'Strength'
output_df['Strength'] = output_df['FSQ'].apply(lambda x: x.split('  ')[1])
# Extract the quantity (same as strength, the part after '  ') from the 'FSQ' column and assign it to a new column 'Quantity'
output_df['Quantity'] = output_df['FSQ'].apply(lambda x: x.split('  ')[2])
```

```python
output_df['Brand'] = output_df['CPDCName']
#removes the brand from the product name
output_df['Product'] = output_df['CPDCName'].str.replace(r'\s?\(.*\)', '', regex=True)
```

```python
output_df['MFRCost'] = output_df['MFRCost'].str.replace('$', '')

```

```python
output_df['Volume_edit'] = output_df['Volume'].str.extract(r'(\d+(\.\d+)?)', expand=False)[0].astype(float)
output_df['Quantity_edit'] = output_df['Quantity'].str.extract(r'(\d+(\.\d+)?)', expand=False)[0].astype(float)
output_df['Volume_num'] = output_df['Volume_edit'] * output_df['Quantity_edit']
```

```python
#filling all the na volumes with quantity value
output_df['Volume'] = output_df['Volume'].apply(lambda x: np.nan if x == '' else x)
output_df['Volume'] = output_df['Volume'].fillna(output_df['Quantity'])
```

```python
output_df['Volume_num'] = output_df['Volume_num'].apply(lambda x: np.nan if x == '' else x)
output_df['Volume_num'] = output_df['Volume_num'].fillna(output_df['Quantity_edit'])
```

```python
output_df.loc[output_df['Volume'].str.contains('liter', case=False, na=False) & 
              ~output_df['Volume'].str.contains('Milliliters', case=False, na=False), 'Volume_num'] = output_df['Volume_edit']*1000
```
